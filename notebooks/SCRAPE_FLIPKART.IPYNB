```python
import pandas as pd
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
from bs4 import BeautifulSoup
import time
import random

options = uc.ChromeOptions()
options.add_argument("--start-maximized")
driver = uc.Chrome(options=options)
print("Chrome launched successfully")

def safe_extract_css(driver, css_selector):
    try:
        element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, css_selector))
        )
        return element.text.strip()
    except:
        return None

def click_element(driver, text):
    try:
        element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f"//*[contains(text(), '{text}')]"))
        )
        driver.execute_script("arguments[0].scrollIntoView(true);", element)
        time.sleep(1)
        element.click()
        print(f"  Clicked: {text}")
        time.sleep(2)
        return True
    except ElementClickInterceptedException:
        try:
            driver.execute_script("arguments[0].click();", element)
            print(f"  JS Clicked: {text}")
            return True
        except:
            print(f"  Click failed: {text}")
            return False
    except Exception as e:
        print(f"  Could not click: {text}")
        return False

def is_visible(driver, text):
    try:
        elements = driver.find_elements(By.XPATH, f"//*[contains(text(), '{text}')]")
        for el in elements:
            if el.is_displayed():
                return True
        return False
    except:
        return False

def click_see_more(driver):
    for label in ["See More", "See more", "READ MORE", "Show More"]:
        try:
            elements = driver.find_elements(By.XPATH, f"//*[contains(text(), '{label}')]")
            for el in elements:
                if el.is_displayed():
                    driver.execute_script("arguments[0].scrollIntoView(true);", el)
                    time.sleep(1)
                    el.click()
                    print(f"  Clicked: {label}")
                    time.sleep(2)
                    return True
        except:
            continue
    print("  No See More button found, continuing")
    return False

def scrape_grid_formation(driver):
    try:
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Find all grid-formation-dynamic containers
        grid_divs = soup.find_all('div', class_='grid-formation-dynamic')
        print(f"  Found {len(grid_divs)} grid-formation-dynamic divs")

        spec_dict = {}

        for grid in grid_divs:
            # Label is in class v1zwn21l, value is in class v1zwn21k
            labels = grid.find_all('div', class_='v1zwn21l')
            values = grid.find_all('div', class_='v1zwn21k')

            for label, value in zip(labels, values):
                label_text = label.get_text(strip=True)
                value_text = value.get_text(strip=True)
                if label_text and value_text:
                    spec_dict[label_text] = value_text
                    print(f"    {label_text}: {value_text}")

        return spec_dict

    except Exception as e:
        print(f"  Grid extraction failed: {e}")
        return {}

def scrape_product(driver, url):
    try:
        driver.get(url)
        time.sleep(random.uniform(3, 5))

        title = safe_extract_css(driver, "h1")
        print(f"  Product Title: {title}")

        # Step 1 - Click All details if Specifications not visible
        if is_visible(driver, "Specifications"):
            print("  Specifications already visible, skipping All details click")
            click_element(driver, "Specifications")
        else:
            print("  Specifications not visible, clicking All details first")
            click_element(driver, "All details")
            time.sleep(2)
            click_element(driver, "Specifications")

        time.sleep(2)

        # Step 2 - Click See More if present
        click_see_more(driver)

        # Step 3 - Scrape all key value pairs from grid
        spec_dict = scrape_grid_formation(driver)

        result = {"Product Title": title}
        result.update(spec_dict)

        return result

    except Exception as e:
        print(f"Failed to scrape {url}: {e}")
        return {"Product Title": None}

# Read CSV
df = pd.read_csv("links_1.csv")

results = []

for idx, row in df.iterrows():
    url = row["Product Link"]
    print(f"\nScraping {idx + 1}/{len(df)}: {url}")
    data = scrape_product(driver, url)
    results.append(data)

driver.quit()

# Merge and save
results_df = pd.DataFrame(results)
final_df = pd.concat([df, results_df], axis=1)

print("\nFinal DataFrame:")
print(final_df.to_string())

final_df.to_csv("out_1.csv", index=False)
print("\nSaved to out_1.csv")
```

```python
import pandas as pd
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
from bs4 import BeautifulSoup
import time
import random
import re

options = uc.ChromeOptions()
options.add_argument("--no-sandbox")
options.add_argument("--start-maximized")
# driver = uc.Chrome(options=options)
driver = uc.Chrome(options=options, version_main=145)
print("Chrome launched successfully")

def safe_extract_css(driver, css_selector):
    try:
        element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, css_selector))
        )
        return element.text.strip()
    except:
        return None

def click_element(driver, text):
    try:
        element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f"//*[contains(text(), '{text}')]"))
        )
        driver.execute_script("arguments[0].scrollIntoView(true);", element)
        time.sleep(1)
        element.click()
        print(f"  Clicked: {text}")
        time.sleep(2)
        return True
    except ElementClickInterceptedException:
        try:
            driver.execute_script("arguments[0].click();", element)
            print(f"  JS Clicked: {text}")
            return True
        except:
            print(f"  Click failed: {text}")
            return False
    except Exception as e:
        print(f"  Could not click: {text}")
        return False

def is_visible(driver, text):
    try:
        elements = driver.find_elements(By.XPATH, f"//*[contains(text(), '{text}')]")
        for el in elements:
            if el.is_displayed():
                return True
        return False
    except:
        return False

def click_see_more(driver):
    for label in ["See More", "See more", "READ MORE", "Show More"]:
        try:
            elements = driver.find_elements(By.XPATH, f"//*[contains(text(), '{label}')]")
            for el in elements:
                if el.is_displayed():
                    driver.execute_script("arguments[0].scrollIntoView(true);", el)
                    time.sleep(1)
                    el.click()
                    print(f"  Clicked: {label}")
                    time.sleep(2)
                    return True
        except:
            continue
    print("  No See More button found, continuing")
    return False

def scrape_price(soup):
    try:
        # Search for any element containing ₹ symbol
        # Filter to find the main price - usually the largest/first one
        price_elements = soup.find_all(
            text=re.compile(r'₹[\d,]+')
        )
        
        prices = []
        for el in price_elements:
            text = el.strip()
            # Extract just the price pattern
            match = re.search(r'₹[\d,]+', text)
            if match:
                price = match.group()
                # Filter out very small numbers which are likely not prices
                number = int(price.replace('₹', '').replace(',', ''))
                if number > 1000:
                    prices.append(price)
        
        if prices:
            # Return first valid price found
            print(f"  Price: {prices[0]}")
            return prices[0]
        return None
    except Exception as e:
        print(f"  Price extraction failed: {e}")
        return None

def scrape_rating(soup):
    try:
        # Rating is always a number like 4.3 or 3.8 between 1 and 5
        rating_elements = soup.find_all(
            text=re.compile(r'^[1-5]\.[0-9]$')
        )
        
        for el in rating_elements:
            text = el.strip()
            if re.match(r'^[1-5]\.[0-9]$', text):
                print(f"  Rating: {text}")
                return text
        
        return None
    except Exception as e:
        print(f"  Rating extraction failed: {e}")
        return None

def scrape_grid_formation(driver):
    try:
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Find all grid-formation-dynamic containers
        grid_divs = soup.find_all('div', class_='grid-formation-dynamic')
        print(f"  Found {len(grid_divs)} grid-formation-dynamic divs")

        spec_dict = {}

        for grid in grid_divs:
            labels = grid.find_all('div', class_='v1zwn21l')
            values = grid.find_all('div', class_='v1zwn21k')

            for label, value in zip(labels, values):
                label_text = label.get_text(strip=True)
                value_text = value.get_text(strip=True)
                if label_text and value_text:
                    spec_dict[label_text] = value_text
                    print(f"    {label_text}: {value_text}")

        return spec_dict, soup

    except Exception as e:
        print(f"  Grid extraction failed: {e}")
        return {}, None

def scrape_product(driver, url):
    try:
        driver.get(url)
        time.sleep(random.uniform(3, 5))

        title = safe_extract_css(driver, "h1")
        print(f"  Product Title: {title}")

        # Step 1 - Click All details if Specifications not visible
        if is_visible(driver, "Specifications"):
            print("  Specifications already visible, skipping All details click")
            click_element(driver, "Specifications")
        else:
            print("  Specifications not visible, clicking All details first")
            click_element(driver, "All details")
            time.sleep(2)
            click_element(driver, "Specifications")

        time.sleep(2)

        # Step 2 - Click See More if present
        click_see_more(driver)

        # Step 3 - Scrape grid and get soup back for price and rating
        spec_dict, soup = scrape_grid_formation(driver)

        # Step 4 - Scrape price and rating from same page source
        price = scrape_price(soup) if soup else None
        rating = scrape_rating(soup) if soup else None

        result = {
            "Product Title": title,
            "Price": price,
            "Rating": rating
        }
        result.update(spec_dict)

        return result

    except Exception as e:
        print(f"Failed to scrape {url}: {e}")
        return {"Product Title": None, "Price": None, "Rating": None}

# Read CSV
df = pd.read_csv("links_2.csv")

results = []

for idx, row in df.iterrows():
    url = row["Product Link"]
    print(f"\nScraping {idx + 1}/{len(df)}: {url}")
    data = scrape_product(driver, url)
    results.append(data)

driver.quit()

# Merge and save
results_df = pd.DataFrame(results)
final_df = pd.concat([df, results_df], axis=1)

print("\nFinal DataFrame:")
print(final_df.to_string())

final_df.to_csv("out_3.csv", index=False)
print("\nSaved to out_2.csv")
```
